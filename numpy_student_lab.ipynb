Section 4 — Vectorization vs Loops

Task 4.1: Loop vs Vectorized

```python
# TODO:
# Create large array X of size 1,000,000

X =  np.random.rand(1_000_000)
X
```

```python
# TODO:
# Normalize using Python loop

# HINT:
# - Time it

# import time
rng = np.random.default_rng(123)
X = rng.standard_normal(1_000_000)

X_mean = np.mean(X)
X_std = np.std(X)

start_time = time.time()

X_normal = np.empty_like(X)
for i in range(X.shape[0]):
    X_normal[i] = (X[i] - X_mean) / X_std

end_time = time.time()
print(f"{round(end_time - start_time, 2)} seconds")

```

```python
# TODO:
# Normalize using vectorization

X_mean = np.mean(X)
X_std = np.std(X)
start_time = time.time()
X_normalized_vectorized = (X - X_mean) / X_std
time_vector = time.time()
X_normalized_vectorized
print(f"{time_vector - start_time} seconds")
```

Why is vectorization faster? : because vectorization performs operation parallely while for loops since it iterates all the elements it is slower


Task 4.2: Pairwise Distance (FAANG Classic)

```python
# TODO:
# Compute pairwise Euclidean distance matrix without loops

# HINT:
# - Use (x - y)^2 expansion
# - Broadcasting is key

def pairwise_distance(X):
  sq = np.sum(X * X, axis = 1, keepdims = True)
  dist = sq + sq.T - 2 * (X @ X.T)
  dist = np.maximum(dist, 0.0)
  return np.sqrt(dist)

rng = np.random.default_rng(3)
Xsmall = rng.standard_normal((200, 10))
Dist = pairwise_distance(Xsmall)
print(Dist.shape)
```

```python
# TODO:
# Implement naive softmax


def softmax_naive(X):
  exp = np.exp(X)
  print(exp)
  return exp/exp.sum(axis = 1, keepdims = True)

Xbig = np.array([[1000.0, 1001.0, 1002.0]])
try:
  softmax_naive(Xbig)
except FloatingPointError as e:
  print('overflowerror :', e)
```

```python
# TODO:
# Fix numerical instability

# HINT:
# - Subtract max per row

def softmax_stable(X):
    Z = X - X.max(axis = 1, keepdims = True)
    exp = np.exp(Z)
    return exp/exp.sum(axis = 1, keepdims = True)

print('softmax_stable :', softmax_stable(Xbig))
print('softmax_rowSum :', softmax_stable(Xbig).sum(axis = 1))
```
